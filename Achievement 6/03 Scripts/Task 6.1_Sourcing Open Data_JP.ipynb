## Check for mixed-type data

```python
# Check for mixed-type columns in df_merged
for col in df_merged.columns.tolist():
  weird = (df_merged[[col]].applymap(type) != df_merged[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_merged[weird]) > 0:
    print (col)
```

```python
# Check for birth year prior to 1920
df_merged_no_dups[df_merged_no_dups["birth year"]< 1920]
```

#### Observations: Records with error make up .005% of total data. Replace 1888 birth year with 1988, and 1900 birth year with 1980 (mean).
